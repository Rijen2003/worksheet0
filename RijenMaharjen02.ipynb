Problem 2 - Data Imputations:
Complete all the following Task:
• Dataset for the Task: "medical_student.csv"
1. Load the provided dataset and import in pandas DataFrame.
2. Check info of the DataFrame and identify column with missing (null) values.
3. For the column with missing values fill the values using various techniques we discussed above. Try
to explain why did you select the particular methods for particular column.
4. Check for any duplicate values present in Dataset and do necessary to manage the duplicate items.
{Hint: dataset.duplicated.sum()}

```python
student_df = pd.read_csv("/content/drive/MyDrive/datassets/medical_students_dataset.csv")
student_df.info()
missing = student_df.isnull().sum()
missing_columns = missing[missing > 0]
print(missing_columns)
```

```python
student_df['Student ID'] = student_df['Student ID'].fillna('unknown')
# since studentd id is unique and cannot be preadicted
student_df['Age']=student_df['Age'].fillna(student_df['Age'].mean())
student_df['Gender']=student_df['Gender'].fillna('unknown')
student_df['Height']=student_df['Height'].fillna(student_df['Height'].mean())
student_df['Weight']=student_df['Weight'].fillna(student_df['Weight'].mean())
student_df['Blood Pressure']=student_df['Blood Pressure'].fillna(student_df['Blood Pressure'].mean())
student_df['Heart Rate']=student_df['Heart Rate'].fillna(student_df['Heart Rate'].mean())
student_df['Cholesterol']=student_df['Cholesterol'].fillna(student_df['Cholesterol'].mean())
student_df['BMI'] = student_df['BMI'].fillna(student_df['BMI'].mean())
student_df['Diabetes']=student_df['Diabetes'].fillna(student_df['Diabetes'].mode()[0])
student_df['Smoking']=student_df['Smoking'].fillna(student_df['Smoking'].mode()[0])
student_df['Blood Type']=student_df['Blood Type'].fillna('unknown')
```

```python
dublicateds = student_df.duplicated().sum()
print(dublicateds)
if(dublicateds>0):
   student_df.drop_duplicates()
   print("clean")
```

Create a DataFrame that is subsetted for the columns ’Name’, ’Pclass’, ’Sex’, ’Age’, ’Fare’, and ’Survived’.

Retain only those rows where ’Pclass’ is equal to 1, representing first-class passengers. What is the mean,
median, maximum value, and minimum value of the ’Fare’ column?

```python
titanic = pd.read_csv("/content/drive/MyDrive/datassets/Titanic-Dataset.csv")
titanic_df = titanic[['Name', 'Pclass', 'Sex', 'Age', 'Fare', 'Survived']]
Pclass_1 = titanic_df[titanic_df['Pclass']==1]
mean = Pclass_1['Fare'].mean()
median = Pclass_1['Fare'].median()
maximum = Pclass_1['Fare'].max()
minimum = Pclass_1['Fare'].min()
print("mean",mean)
print("median",median)
print("maximum",maximum)
print("minimum",minimum)
```

Problem - 4:

Compare the mean survival rates (’Survived’) for the different groups in the ’Sex’ column. Draw a visual-
ization to show how the survival distributions vary by gender.

```python
import matplotlib.pyplot as plt
import seaborn as sns
mean_survival = titanic.groupby('Sex')['Survived'].mean()
print(mean_survival)
sns.barplot(x=mean_survival.index, y=mean_survival.values)
plt.xlabel('Gender')
plt.ylabel('Mean Survival Rate')
plt.title('Mean Survival Rate by Gender')
plt.show()
```

Problem - 6{Optional}:
Show how the survival rates (’Survived’) vary by age group and passenger class (’Pclass’). Break up the
’Age’ column into five quantiles in your DataFrame, and then compare the means of ’Survived’ by class
and age group. Draw a visualization using a any plotting library to represent this graphically.

```python
t_read['Age_Group'] = pd.qcut(t_read['Age'], q=5)
mean_Sur = t_read.groupby(['Pclass', 'Age_Group'])['Survived'].mean().unstack()
print(mean_Sur)
mean_Sur.plot(kind='bar', figsize=(10, 6), colormap='viridis')
plt.xlabel('Age Group')
plt.ylabel('Mean Survival Rate')
plt.title('Mean Survival Rate by Age Group and Passenger Class')
plt.show()
```
